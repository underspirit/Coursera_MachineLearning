### 构建模型
优化的目标函数：
$$J(x^{(1)},...,x^{(n_m)},\theta^{1},...,\theta^{n_u}) = 
\frac{1}{2} \sum_{(i,j) : r(i,j)=1} \left( (\theta^{(j)})^T x^{(i)} - y^{(i,j)} \right)^2 
+ \frac{\lambda}{2} \sum_{i=1}^{n_m} \sum_{k=1}^{n} (x_k^{(i)})^2
+ \frac{\lambda}{2} \sum_{j=1}^{n_u} \sum_{k=1}^{n} (\theta_k^{(j)})^2 $$   
其中$n_u$为用户数量，$n_m$为电影数量，$n$为电影的特征维度，$r(i,j)=1$表示用户i给电影j评过分，$y^{(i,j)}$表示用户i给电影j的打分。   

对目标函数求偏导数,梯度下降的迭代过程如下：
$$ x_k^{(i)} := x_k^{(i)} - \alpha \left( \sum_{j:r(i,j)=1} \left( (\theta^{(j)})^T x^{(i)}  - y^{(i,j)} \right) \theta_k^{(j)} + \lambda x_k^{(i)}   \right) $$
$$ \theta_k^{(j)} := \theta_k^{(j)} - \alpha \left( \sum_{i:r(i,j)=1} \left( (\theta^{(j)})^T x^{(i)}  - y^{(i,j)} \right) x_k^{(i)} + \lambda \theta_k^{(j)}   \right) $$

```python
x = theano.shared(np.zeros([2,2], dtype='float32'), name='x', ) # (1682, 100)
theta = theano.shared(np.zeros([2,2], dtype='float32'), name='theta') # (943, 100)
y = T.matrix(name='y') # (1682, 943)
r = T.matrix(name='r') # (1682, 943)
_lambda =  T.scalar(name='lambda')

ratings = T.dot(x, theta.T) # (1682, 943) 计算打分
deviation = ratings * r - y # 乘以r矩阵，未打分的为0

# 损失函数
cost = T.sum((deviation) ** 2) / 2 + (_lambda / 2) * T.sum(x ** 2) + (_lambda / 2) * T.sum(theta ** 2) 

computeCost = theano.function([y, r, _lambda], cost)
computeRatings = theano.function([], ratings)
```

```python
# 手动公式求导
grad_x = T.dot(deviation, theta)  + _lambda * x 
grad_theta = T.dot(deviation.T, x)  + _lambda * theta
computeGrad2 = theano.function([y, r, _lambda], [grad_x, grad_theta])

# 自动求导
gradAuto_x = T.grad(cost , x)
gradAuto_theta = T.grad(cost , theta)
computeGrad = theano.function([y, r, _lambda], [gradAuto_x, gradAuto_theta])
```

###  梯度检验

```python
def computeNumericalGradient(X, Theta, Y, R, _lambda):
    numgrad_X = np.zeros_like(X)
    numgrad_Theta = np.zeros_like(Theta)
    perturb_X = np.zeros_like(X).astype('float32')
    perturb_Theta = np.zeros_like(Theta).astype('float32')
    e = 1e-4
    for (i, j), _ in np.ndenumerate(X):
        perturb_X[i, j] = e
        x.set_value(X - perturb_X)
        loss1 = computeCost(Y, R, _lambda)
        x.set_value(X + perturb_X)
        loss2 = computeCost(Y, R, _lambda)
        # Compute Numerical Gradient
        numgrad_X[i, j] = (loss2 - loss1) / (2 * e)
        perturb_X[i, j] = 0        
        
    for (i, j), _ in np.ndenumerate(Theta):
        perturb_Theta[i, j] = e
        theta.set_value(Theta - perturb_Theta)
        loss1 = computeCost(Y, R, _lambda)
        theta.set_value(Theta + perturb_Theta)
        loss2 = computeCost(Y, R, _lambda)
        # Compute Numerical Gradient
        numgrad_Theta[i, j] = (loss2 - loss1) / (2 * e)
        perturb_Theta[i, j] = 0   
        
    return numgrad_X, numgrad_Theta

def checkNNGradients(lambdaVal):
    # Create small problem
    X_t = np.random.rand(4, 3).astype('float32')
    Theta_t = np.random.rand(5, 3).astype('float32')
    
    # Zap out most entries
    Y = np.dot(X_t, Theta_t.T) # (4, 5)
    idx = np.nonzero(np.random.rand(Y.shape[0], Y.shape[1]) > 0.5) # 未评分的元素
    Y[idx] = 0 
    R = np.ones_like(Y) # 未评分标记
    R[idx] = 0

    # Run Gradient Checking
    X = np.random.randn(X_t.shape[0], X_t.shape[1]).astype('float32')
    Theta = np.random.randn(Theta_t.shape[0], Theta_t.shape[1]).astype('float32')
    num_users = Y.shape[1]
    num_movies = Y.shape[0]
    num_features = Theta_t.shape[1]
    
    x.set_value(X.astype('float32'))
    theta.set_value(Theta.astype('float32'))
    grad_X, grad_Theta = computeGrad2(Y, R, lambdaVal)
    grad_all = np.concatenate([grad_X.flatten(), grad_Theta.flatten()])
    numgrad_X, numgrad_Theta = computeNumericalGradient(X , Theta, Y, R, lambdaVal)
    numgrad_all = np.concatenate([numgrad_X.flatten(), numgrad_Theta.flatten()])

    # format numpy array print
    float_formatter = lambda x: "%.4f" % x
    np.set_printoptions(formatter={'float_kind':float_formatter})
    grad_all.shape = -1, 1
    numgrad_all.shape = -1, 1
    print np.hstack((grad_all, numgrad_all))
    diff = np.linalg.norm(numgrad_all - grad_all) / np.linalg.norm(numgrad_all + grad_all)
    print 'Relative Difference: %g' % diff
  
    
checkNNGradients(0)
```

### 添加自己为一个新用户，并对部分电影进行评分

```python
# Initialize my ratings
my_ratings = np.zeros(1682)

# Check the file movie_idx.txt for id of each movie in our dataset
# For example, Toy Story (1995) has ID 1, so to rate it "4", you can set
my_ratings[0] = 4;
my_ratings[87] = 4;
my_ratings[90] = 3.6;
my_ratings[95] = 4.2;
my_ratings[142] = 2.8;
my_ratings[182] = 3.8;
my_ratings[195] = 3.8;
my_ratings[196] = 3.2;
my_ratings[199] = 4.2;
my_ratings[209] = 4.0;
my_ratings[224] = 3.6;
my_ratings[249] = 4.2;
my_ratings[251] = 4.0;
my_ratings[254] = 3.0;
my_ratings[312] = 4.2;

for i, rating in enumerate(my_ratings):
    if rating > 0:
        print 'Rated %d for %s\n' % (my_ratings[i], movieList[i])
```

```python
# 标准化评分矩阵，每部电影均值为0，未评分电影（0）则为该电影的
def normalizeRatings(Y, R):
    sumY = Y.sum(axis=1)
    Ymean = sumY / R.sum(axis=1)
    Ynorm = Y - Ymean.reshape(-1, 1)
    Ynorm = Ynorm * R
    return Ynorm, Ymean
```

```python
# 将自己添加到评分矩阵
Y_p = np.concatenate((my_ratings.reshape(-1, 1), Y), 1).astype('float32')
R_p = np.concatenate(((my_ratings != 0).reshape(-1, 1), R), 1).astype('float32')
Ynorm, Ymean = normalizeRatings(Y_p, R_p)
num_users = Y_p.shape[1]
num_movies = Y_p.shape[0]
num_features = 10

lambda_val = np.float32(10)
```

### 使用scipy.optimize优化算法训练

```python
# 计算指定theta时的cost,使用optimize的函数时,传入的待优化参数是一个一维向量,所以要转成二维列向量,并且使用fmin时还需要转成float32类型
def train_fn_cost(theta_all, lambdaVal):
    x.set_value(theta_all[:num_movies * num_features].reshape([num_movies, num_features]).astype('float32'))
    theta.set_value(theta_all[num_movies * num_features :].reshape([num_users, num_features]).astype('float32'))
    return computeCost(Ynorm, R_p, lambdaVal)

# 计算指定theta时的梯度
def train_fn_grad(theta_all, lambdaVal):
    x.set_value(theta_all[:num_movies * num_features].reshape([num_movies, num_features]).astype('float32'))
    theta.set_value(theta_all[num_movies * num_features :].reshape([num_users, num_features]).astype('float32'))
    grad_X, grad_Theta  = computeGrad(Ynorm, R_p, lambdaVal)
    return np.hstack([grad_X.flatten(), grad_Theta.flatten()]) # 梯度也要转为一维向量


def fitModel(lambdaVal):
    # 使用共轭梯度法优化
    thetaValue = optimize.fmin_cg( # or fmin_bfgs
        f=train_fn_cost,
        x0=np.random.randn(num_movies*num_features + num_users * num_features).astype('float32'), # same as matlab
        fprime=train_fn_grad,
        args=(lambdaVal,),
        disp=1,
        maxiter=100
    )
    return thetaValue
```

```python
import time
start_time = time.time()
theta_all = fitModel(lambda_val)
end_time = time.time()
print 'training cost %.2f s' % (end_time - start_time)
#theta.set_value(thetaValue.astype('float32'))
#print 'training set accuracy is %.2f ' % (calcAccur(10, 25)) # 100%
```

### 为自己推荐电影

```python
# 计算最终的评分矩阵
x.set_value(theta_all[:num_movies * num_features].reshape([num_movies, num_features]).astype('float32'))
theta.set_value(theta_all[num_movies * num_features :].reshape([num_users, num_features]).astype('float32'))

ratings = computeRatings() 
my_predictions = ratings[:, 0] + Ymean # 我的打分
my_predictions_sortedIdx =  np.argsort(my_predictions)[::-1] # 分数从高到低排序
userCount = np.sum(R_p, 1) # 每部电影的打分人数
```

```python
# 为我推荐10部电影
n = 10
k = 0
for i in range(len(my_predictions)):
    idx = my_predictions_sortedIdx[i]
    if k == 10:
        break
    if userCount[idx] < 15: # 打分人数过少的电影不推荐
        continue
    k += 1
    print 'Predicting rating %.1f for movie %s\n' % (my_predictions[idx], movieList[idx])
```
